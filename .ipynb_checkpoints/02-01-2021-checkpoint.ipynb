```python
#tsv --> tab separated values
df_tsv = pd.read_csv('pandas data-set/test.tsv',sep="\t")
df_tsv
```

```python
## without separater
df_tsv = pd.read_table('pandas data-set/ex1.csv')
df_tsv
```

```python
df_tsv = pd.read_table('pandas data-set/ex1.csv',sep=',')
df_tsv
```

```python
## 'infer' --> by default 1st row will be the header
df_tsv = pd.read_table('pandas data-set/ex1.csv',sep=',',header='infer')  
df_tsv
```

```python
df_tsv = pd.read_table('pandas data-set/ex1.csv',sep=',',header=None)  
df_tsv
```

```python
## names will provide a column name 
df_tsv = pd.read_table('pandas data-set/ex1.csv',sep=',',header=None
                      ,names=['hello','Kumaru','machan','loosu','good'])  
df_tsv
```

```python
## 1st column became the row index since i gave only 4 column names 
df_tsv = pd.read_table('pandas data-set/ex1.csv',sep=',',header=None
                      ,names=['hello','Kumaru','machan','loosu'])  
df_tsv
```

```python
## Now 2 columns became the row index
df_tsv = pd.read_table('pandas data-set/ex1.csv',sep=','
                      ,names=['hello','Kumaru','machan'])  
df_tsv
```

```python
## use columns (usecols) --> can pull out a specif column
df_tsv = pd.read_table('pandas data-set/ex1.csv',sep=',',header=None
                      ,names=['hello','Kumaru','machan','loosu','good']
                      ,usecols=['Kumaru'])  
df_tsv
```

```python
'''
index_col --> uniquely identifies the record 
find 4, which is in (one,b,value2)
'''

df = pd.read_csv('pandas data-set/csv_mindex.csv',
                 index_col=['key1', 'key2'])
df
```
